#### Remark:
- It can be shown easily that the weighted average life (as discussed in
wiki) is the sum of time-varying principals/initial principal.

- Based on the formula in **Amortization formula**, https://en.wikipedia.org/wiki/Amortization_calculator, we can code the
**weighted average duration** and **total interest** in the most ideal situation (terimated at maturity, with no penalty, no settlment, ...) as


```python
### N    - terms in months
### rate - monthly interest rate ~ annual interest rate/12

def Weighted_Average_Duration(N, rate):
    if np.isnan(N) or np.isnan(i): return np.nan
    
    p = 1.0-(np.exp(np.arange(N+1)*np.log(1+rate))-1)/(np.exp(N*np.log(1+rate))-1)
    # principals time series
    return np.sum(p)

def Total_Cumulative_Interest(N, rate):
    if np.isnan(N) or np.isnan(rate): return np.nan

    fac = np.exp(-N*np.log(1+rate))
    return N*rate/(1-fac)-1
    
```

### Pooled Models vs Time Series Models
- The column 'issue_d' (**issuance date**) is in the year-month format.
- When we treat it as a categorical column representing differnt sub-groups of data points, the **pooled model**, a jargon of statisticians, refers to
the model trained by lumping all data of different issuance dates together.
- In comparison, a time series ML model trains sequentially the data points with different issuance dates and the model evolve w.r.t. time as new data points come in.
- The pooled model summarizes the historical data, which can be used as a 
baseline of future time-series ML models.
- On the other hand, the sequential (batch) models micmic the real-world
scenario as the real-world data comes in month-after-month.
